```python
import numpy as np

from blackjack_env import BlackjackEnv
from utils import monte_carlo_on_policy, plot_blackjack_monte_carlo_es
```

## **Example**

```python
env = BlackjackEnv()
```

```python
policy, Q = monte_carlo_on_policy(env, epsilon=0.15, init_policy=None, max_episodes=500000, verbose=True) # We will start with uniform policy
```

```python
# Since the policy is stochastic, we convert it to deterministic by taking the argmax for plotting.

greedy_policy = np.argmax(policy, axis=-1)
plot_blackjack_monte_carlo_es(greedy_policy, Q)
```
