## **Example**

```python
env = BlackjackEnv()

policy = np.ones(shape=(22, 11, 2), dtype=int) * env.HIT # alway HIT
policy[[20, 21], :, :] = env.STICK # STICK when player_sum is 20 or 21
```

```python
policy, Q = monte_carlo_es(env, init_policy=policy, max_episodes=2000000, verbose=True)
```
